####  Given a CSV file with student names and scores, find the student with the highest score 

```python
   

# CSV file name
csv_file_name = "student_scores.csv"

# Initialize variables to track the highest score and the corresponding student
highest_score = -1  # Initialize with a value lower than any possible score
top_student = None

# Open the CSV file and read it line by line
with open(csv_file_name, mode="r") as csv_file:
    # Skip the header line
    next(csv_file)
    
    for line in csv_file:
        parts = line.strip().split(',')
        if len(parts) == 2:
            student_name = parts[0]
            score = int(parts[1])
            
            if score > highest_score:
                highest_score = score
                top_student = student_name

if top_student:
    print(f"The student with the highest score is {top_student} with a score of {highest_score}.")
else:
    print("No student data found in the CSV file.")
```

#### Implement a program that reads a text file and counts the number of words and lines in it

```python
read_file = 'shep_n_drake.txt'

def reading_lines(content):
    now = len(content.split())
    nol = len(content.splitlines())
    
    return now, nol
    
    


try:
    with open(read_file, 'r') as readint_lines:
        content = readint_lines.read()
        now,nol = reading_lines(content)
        print("Number of words in the file are : ", now, " & number of lines are : ", nol)
        
        
except FileNotFoundError:
    print("File not found")
except Exception as e:
    print("Error happened",e)
    
```
